```python
import pandas as pd
import numpy as np
dados = pd.read_csv('https://gist.githubusercontent.com/guilhermesilveira/1b7d5475863c15f484ac495bd70975cf/raw/16aff7a0aee67e7c100a2a48b676a2d2d142f646/projects.csv')
```

```python
dados.head()
```

```python
colunas = {
    'unfinished': 'nao_finalizado',
    'expected_hours': 'horas_experadas',
    'price': 'preco'
}
dados = dados.rename(columns=colunas)
dados.head()
```

```python
trocar = {
    0:1,
    1:0
}

dados['finalizado'] = dados['nao_finalizado'].map(trocar)
dados.head()
```

```python
dados.tail()
```

```python
import seaborn as sns
sns.scatterplot(data=dados, x = 'horas_experadas', y = 'preco')
```

```python
sns.scatterplot(data=dados, x = 'horas_experadas', y = 'preco', hue = 'finalizado')
```

```python
sns.relplot(data=dados, x = 'horas_experadas', y = 'preco', col = 'finalizado', hue='finalizado')
```

```python
x = dados[['horas_experadas','preco']]
y = dados['finalizado']
from sklearn.model_selection import train_test_split
SEED = 5
np.random.seed(SEED)
treino_x, teste_x, treino_y, teste_y = train_test_split(x,y, test_size=0.25, stratify=y)

from sklearn.svm import LinearSVC
modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

from sklearn.metrics import accuracy_score
accuracy_score(teste_y, previsoes)
print('Treinamos Com %d e Testamos com %d' % (len(treino_x),len(teste_x)))
print('Acuracia foi %.2f%%' % (accuracy_score(teste_y,previsoes)*100))
```

```python
from sklearn.metrics import accuracy_score
print('Acuracia algoritmo de base line foi %.2f%%' % (accuracy_score(teste_y,previsoes_de_base)*100))
```

```python
from sklearn.preprocessing import StandardScaler
x = dados[['horas_experadas','preco']]
y = dados['finalizado']
from sklearn.model_selection import train_test_split
SEED = 5
np.random.seed(SEED)
raw_treino_x, raw_teste_x, treino_y, teste_y = train_test_split(x,y, test_size=0.25, stratify=y)

scaler = StandardScaler()
scaler.fit(raw_treino_x)
treino_x = scaler.transform(raw_treino_x)
teste_x = scaler.transform(raw_teste_x)

from sklearn.svm import SVC
modelo = SVC(gamma='auto')
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

from sklearn.metrics import accuracy_score
accuracy_score(teste_y, previsoes)
print('Treinamos Com %d e Testamos com %d' % (len(treino_x),len(teste_x)))
print('Acuracia foi %.2f%%' % (accuracy_score(teste_y,previsoes)*100))
```

```python
data_x = teste_x[:,0]
data_y = teste_x[:,1]
x_min = data_x.min()
x_max = data_x.max()
y_min = data_y.min()
y_max = data_y.max()
print(x_min, x_max,y_min,y_max)

pixel = 100
eixo_x = np.arange(x_min, x_max, (x_max-x_min) / pixel)
eixo_y = np.arange(y_min, y_max, (y_max-y_min) / pixel)

xx,yy = np.meshgrid(eixo_x,eixo_y)
pontos = np.c_[xx.ravel(),yy.ravel()]
pontos

Z = modelo.predict(pontos)

Z.shape

Z = Z.reshape(xx.shape)
Z

import matplotlib.pyplot as plt
plt.contourf(xx,yy,Z, alpha = 0.3)
plt.scatter(teste_x.horas_experadas, teste_x.preco, c = teste_y, s=1)
```
